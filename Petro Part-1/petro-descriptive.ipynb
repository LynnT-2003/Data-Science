## Sales by pump number

```python
sales_by_pump = journal_detail.groupby("pumpnumber")["total"].sum().reset_index()
sales_by_pump = sales_by_pump.sort_values(by="total", ascending=False)
```

```python
sales_by_pump
```

```python
# plot the sales data by pump number
sns.barplot(x="pumpnumber", y="total", data=sales_by_pump)
plt.title("Sales by Pump Number")
plt.xlabel("Pump Number")
plt.ylabel("Total Sales")
plt.show()
```

Pump number 10 has a significantly lower sales compared to all the other pumps.
There are possible reasons for this:
1. The pump could've been faulty for some time (what is it pumping?)
2. Traffic into the stations are not equally distributed
3. The location of the pump (for example, an isolated petro station with only pump number 10 installed)

# Total sales over hour of day


```python
# convert journaldatetime column to datetime format
dfMerged['journaldatetime'] = pd.to_datetime(dfMerged['journaldatetime'])

# group the data by hour and calculate the total sales for each hour
sales_by_hour = dfMerged.groupby(dfMerged['journaldatetime'].dt.hour)['total_x'].sum()

# create a line plot of the total sales by hour
sns.barplot(x=sales_by_hour.index, y=sales_by_hour.values)
plt.title('Total Sales by Hour')
plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.show()
```

There is an obvious trend in total sales by hour<br/>
0-4am is the time when most people are inactive on the roads (sleeping hours) <br/>
5-6am is still considered too early to get up and get on the road for some people. <br/>
16.00 and 18.00 are the most popular since they are times people get off their office hours (rush hours) <br/>
And as it gets later, there are more people indoors and less people on the roads who need to fill up their gas tanks at petro stations <br/>

# Sales over day of the week

```python
# Convert journaldatetime column to datetime format
dfMerged['journaldatetime'] = pd.to_datetime(dfMerged['journaldatetime'])

# Create a new column for the day of the week
dfMerged['day_of_week'] = dfMerged['journaldatetime'].dt.day_name()

# Group the data by day of the week and calculate the total sales for each day
sales_by_day = dfMerged.groupby('day_of_week')['total_x'].sum()

# Create a bar plot of the total sales by day of the week
sns.barplot(x=sales_by_day.index, y=sales_by_day.values)
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.show()
```

Barchart wasn't that effective, so lets do a correlation analysis

```python
import numpy as np
```

```python
# group the data by day of the week and calculate the total sales for each day
sales_by_day = dfMerged.groupby(dfMerged['journaldatetime'].dt.day_name())['total_x'].sum()

# create a list of day names in order
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# calculate the correlation coefficient between the total sales and the day of the week
corr = np.corrcoef(sales_by_day, range(7))[0][1]

# print the correlation coefficient
print(f"Correlation between sales and day of the week: {corr:.2f}")
```

There definitely is some correlation between sales and day of the week, lets do a line plot instead

```python
# Create a line plot of the total sales by day of the week
sns.lineplot(x=sales_by_day.index, y=sales_by_day.values)
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.show()
```

There is significantly less sales on Sunday. <br/>
Sales are lowest on weekends for obvious reasons. <br/>
Sales are highest on Fridays because its a Friday (Friday Nights and weekends are coming)

# Sales trend for each product per year

```python
# Convert journaldatetime column to datetime format
dfMerged['journaldatetime'] = pd.to_datetime(dfMerged['journaldatetime'])

# Extract the year from the journaldatetime column and create a new column for it
dfMerged['year'] = dfMerged['journaldatetime'].dt.year

# Group the data by year and product name and calculate the total sales for each group
sales_by_year_product = dfMerged.groupby(['year', 'item_name'])['total_x'].sum().reset_index()

# Create a line plot for each product
for product in sales_by_year_product['item_name'].unique():
    df = sales_by_year_product[sales_by_year_product['item_name'] == product]
    sns.lineplot(x='year', y='total_x', data=df, label=product)

plt.title('Sales Trend for Each Product per Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(loc='upper left')
plt.show()
```

Sales for all products went down during 2018-2019 with GASOHOL91 with the greatest decrease (steepest gradient) <br/>
A more deeper insight to 'Sales Trend over Time' <br/>
E20, Diesel Plus, and Safe Drop Cash have always been the least sold
